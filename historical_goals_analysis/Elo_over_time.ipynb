```python
import datetime as dt
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/team_elos_over_time.csv', parse_dates=['date'], dtype={'rating': int})
```

```python
df = df[['date', 'team', 'rating']]
df
```

```python
def plot_country_rating(df, countries):
    fig, ax = plt.subplots(facecolor='white')
    for country in countries:
        df_country = df[df['team'] == country].set_index('date')
        resampled = df_country.resample('d').pad()
        ax.plot(resampled['rating'], label=country)
        ax.legend()
    ax.set_xlabel('Date')
    ax.set_ylabel('Elo rating')
    plt.show()
    
plot_country_rating(df, ['France', 'Belgium', 'Germany'])
```
